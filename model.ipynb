🏷️ Encoding Categorical Features
In this step, we transform the categorical (string) columns in our input DataFrame into numerical format using one-hot encoding. This is essential for many machine learning models that require numerical input.

🔄 What We Did:
Applied pd.get_dummies() to convert the following categorical columns into binary indicator variables:

Platform

Hashtag

Content_Type

Region

This process creates a new column for each unique category in those features, allowing models to interpret the categorical information numerically.

✅ Result:
The resulting DataFrame, inputs_encoded, contains the original numeric features (if any) and the newly created dummy variables for each categorical feature.

We display the first 10 rows of the transformed dataset to verify the changes.

```python



inputs_1 = pd.get_dummies(inputs, columns=['Platform','Hashtag', 'Content_Type', 'Region'])
inputs_encoded=inputs_1.drop(['Hashtag_Challenge'], axis='columns')

inputs_encoded.head(10)

```

🔢 Label Encoding the Target Variable
In this step, we convert the categorical target variable (Engagement_Level) into numeric form using LabelEncoder from scikit-learn.

✅ Why Label Encoding?
Most machine learning models require the target variable to be numeric. LabelEncoder assigns a unique integer to each class label, which is suitable for ordinal or classification targets.

🔄 What We Did:
Initialized and applied LabelEncoder to the target Series.

Created a new DataFrame, target_df, that includes:

The original Engagement_Level values

The newly encoded Engagement_Level_n values

Dropped the original string column (Engagement_Level) to retain only the numeric labels in target_final.

📦 Result:
The final target DataFrame target_final contains only numeric labels and is ready to be used for model training.

We displayed the first 20 rows to verify the encoding.

```python

le = LabelEncoder()
encoded_labels = le.fit_transform(target)

# Add to DataFrame for clarity
target_df = pd.DataFrame({
    'Engagement_Level': target,
    'Engagement_Level_n': encoded_labels
})

# Display result
target_df.head(20)

target_final=target_df.drop(['Engagement_Level'], axis='columns')

target_final.head(20)
```

```python
inputs_encoded.head(20)
```

📊 Model Comparison Using Grid Search and Train-Test Split
In this section, we compare multiple classification models using GridSearchCV for hyperparameter tuning and evaluate each model's performance on a held-out test set.

✅ Steps Performed:
Defined multiple classifiers:

Support Vector Machine (SVC)

Random Forest

Logistic Regression

Decision Tree

Specified a range of hyperparameters for each model to tune using cross-validation.

Used GridSearchCV with 5-fold cross-validation on the training set to find the best parameters for each model.

Evaluated each best model on the test set using .score() to measure accuracy.

📈 Output:
The final table displays:

Model name

Best cross-validation score on the training data

Test score on unseen data

Best hyperparameters selected by GridSearchCV

This approach helps identify the model that not only fits the training data well but also generalizes best to unseen data.

```python



X = inputs_encoded
y = target_final

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y.values.ravel(),  
    test_size=0.2,
    stratify=target_final,
    random_state=42
)


# Output the shapes of the resulting splits
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Define model parameters
model_params = {
    'svm': {
        'model': svm.SVC(),
        'params': {
            'C': [1],
            'kernel': ['linear'],
            'gamma': ['scale'],
            'class_weight': [None]
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [50],
            'max_depth': [10],
            'min_samples_split': [2],
            'min_samples_leaf': [1],
            'max_features': ['sqrt'],
            'bootstrap': [True]
        }
    },
    'Logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1],
            'penalty': ['l2'],
            'solver': ['liblinear'],
            'max_iter': [100],
            'class_weight': [None]
        }
    },
    'Decision_tree': {
        'model': tree.DecisionTreeClassifier(),
        'params': {
            'criterion': ['gini'],
            'max_depth': [10],
            'min_samples_split': [2],
            'min_samples_leaf': [1],
            'max_features': [None]
        }
    }
}

# Store results
scores = []

# Run RandomizedSearchCV for each model
for model_name, mp in model_params.items():
    clf = RandomizedSearchCV(mp['model'], mp['params'], n_iter=1, cv=5, random_state=42, return_train_score=False)
    clf.fit(X_train, y_train)

    test_score = clf.score(X_test, y_test)

    scores.append({
        'model': model_name,
        'best_cv_score (train)': clf.best_score_,
        'test_score': test_score,
        'best_params': clf.best_params_
    })

# Create DataFrame of scores
score_df = pd.DataFrame(scores, columns=['model', 'best_cv_score (train)', 'test_score', 'best_params'])
score_df

```
